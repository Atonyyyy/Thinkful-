```python
plt.xlabel("Age")
plt.ylabel("Index")
plt.scatter(x=df["age"],y=df.index)
plt.show()
## Data shows that age of heart disease present is more common between ages of 40-65.
```

```python
plt.title("Heart Disease")
plt.scatter(x=df["chol"],y=df.index)
plt.xlabel("Cholesterol")
plt.ylabel("Index")
plt.show()

## Most people in this data set with heart disease present has cholesterol levels of 220 - 300.
```

```python
plt.xlabel("Age")
plt.title("Heart disease is more present in people aged 40+ with Cholesterol levels of 220+")
plt.ylabel("Cholesterol")
plt.scatter(x=df["age"],y=df["chol"])
plt.show()

```

```python
x=df["age"]
plt.hist(x,color="red", alpha =.5)
plt.title("Age Histogram")


plt.show()

#Users are more affected with heart disease within age range of 55-65.
```
